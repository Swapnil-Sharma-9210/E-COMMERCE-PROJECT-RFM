```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Superstore.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
```

```python
df.drop('Row ID',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
```

```python
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
```

```python
df['order_year'] = df['Order Date'].dt.year
df['order_month'] = df['Order Date'].dt.month
df['order_day'] = df['Order Date'].dt.day
```

```python
df['Sales'] = df['Sales'].astype(float)
df['Cost'] = df['Sales'] - df['Profit']
```

```python
df.info()
```

```python
print('The most recent Order Date is:',df['Order Date'].max())
print('The first Order Date is:',df['Order Date'].min())
```

```python
df_recency = df.groupby(by='Customer Name',as_index=False)['Order Date'].max()
df_recency.columns = ['CustomerName', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(lambda x: (recent_date - x).days)
df_recency.head()
```

```python
frequency_df = df.groupby(by=['Customer Name'], as_index=False)['Order Date'].count()
frequency_df.columns = ['CustomerName', 'Frequency']
frequency_df.head()
```

```python
df['Total'] = df['Sales']*df['Quantity']
monetary_df = df.groupby(by='Customer Name', as_index=False)['Total'].sum()
monetary_df.columns = ['CustomerName', 'Monetary']
monetary_df.head()
```

```python
rf_df = df_recency.merge(frequency_df, on='CustomerName')
rfm_df = rf_df.merge(monetary_df, on='CustomerName').drop(columns='LastPurchaseDate')
rfm_df.head()
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)
```

```python
# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['M_rank']/rfm_df['M_rank'].max())*100
```

```python
rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
```

```python
rfm_df.head()
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.28*rfm_df['F_rank_norm']+0.57*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df.head()
```

```python
rfm_df['RFM_Score'].describe()
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] > 4.5,"Top Customers",
                            (np.where(rfm_df['RFM_Score'] > 4,"High value Customer",
                            (np.where(rfm_df['RFM_Score'] > 3,"Medium Value Customer",
                             np.where(rfm_df['RFM_Score'] > 1.5,"Low Value Customers", 'Lost Customers'))))))                                                                            
```

```python
rfm_df.head()
```

```python
customer_segment_counts = rfm_df['Customer_segment'].value_counts()

plt.figure(figsize=(15, 10)) 
plt.bar(customer_segment_counts.index, customer_segment_counts.values)
plt.xlabel('Customer Segment')
plt.ylabel('Count')
plt.title('Customer Segment Distribution')


plt.show()
```
